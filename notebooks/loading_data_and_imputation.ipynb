## Missing value first steps

Looking directly at the return series, we have some data cleaning to do. 

The first thing we observe for is that Fund F is Null for all dates on and before May 2nd of 2021. This will not pose a serious problem for the present analysis since the fund manager did not have allocation to Fund F until June of 2021.

What we will do is replace the null values of Fund F with 0 to avoid messing with future steps which will rely on null detection.

```python
fund_returns = returns['Fund Total Returns (CAD)']
index_returns = returns['Index Total Returns (CAD)']
```

```python
# todo find better way to suppress copy warning
pd.set_option('mode.chained_assignment', None)

fund_returns['Fund F'].fillna(0, inplace=True)
```

## Addressing holiday fund returns

It is likely that the reason for null values is due to the Canadian holiday schedule. A problem arises wherein if the underlying assets on which a fund is based are traded on the Canadian holiday (i.e. on foreign exchanges--consider US equities being traded on Canada day), then having a 0% return on the Canadian holiday will likely systemically over-represent tracking error for the fund. We should not simply replace NAN with 0. 

We will accordingly use a two-pronged approach for filling missing values. We first do a check if the underlying was traded (non-zero return as a proxy). If not traded, we fill with 0, as this will track the underlying assets appropriately. If traded, we will use a geometric average of the following day's trade and 0% return, such that the information of the fund movement is spread across the holiday and the following day.

Looking at the null holiday values, we see a situation where even though fund D, F, and fund J have values present, they are unlikely to be true values since they are either 0, very small values, or repeating very specific numbers. 

We will thus treat these funds as if they were null for these holiday values with respect to how we treat them from a null-filling standpoint.

```python
# show holiday NANs
fund_returns[fund_returns.isna().any(axis=1)]
```

```python
# save dates that we need to modify
dates_to_modify = list(fund_returns[fund_returns.isna().any(axis=1)].index)
```

```python
# funds and their corresponding underlying asset types
fund_index_lookup = {
    "Fund A": "Canada equities",
    "Fund B": "Canada equities",
    "Fund C": "US equities",
    "Fund D": "US equities",
    "Fund E": "EAFE equities",
    "Fund F": "EAFE equities",
    "Fund G": "EM equities",
    "Fund H": "EM equities",
    "Fund I": "Canada Broad FI",
    "Fund J": "Canada Broad FI",
    "Fund K": "Canada Corps FI",
    "Fund L": "Canada Corps FI"
}
```

```python
fund_returns_copy = fund_returns.copy()

# get the necessary time stamps for each row

fund_returns_copy["t_0"] = list(fund_returns_copy.reset_index()["index"])
fund_returns_copy["t_minus_1"] = fund_returns_copy["t_0"].shift(-1)
fund_returns_copy["t_plus_1"] = fund_returns_copy["t_0"].shift(1)
```

## Complications

The existance of double-stacked holidays introduces some extra complexity, need to address these cases individually.

```python
fund_returns_copy["one_value_missing_flag"] = fund_returns_copy.apply(
    lambda row: 1 if (row["t_0"] in dates_to_modify) & (
                      row["t_minus_1"] not in dates_to_modify) &(
                      row["t_plus_1"] not in dates_to_modify)
    else 0, axis=1)

fund_returns_copy["two_values_missing_flag"] = fund_returns_copy.apply(
    lambda row: 1 if (row["t_0"] in dates_to_modify) & (
                      row["t_minus_1"] in dates_to_modify) & (
                      row["t_plus_1"] not in dates_to_modify) 
    else 0, axis=1)

assert (fund_returns_copy["one_value_missing_flag"].sum() +
        fund_returns_copy["two_values_missing_flag"].sum() * 2) == len(
    dates_to_modify)
```

```python
fund_returns_copy.loc[fund_returns_copy["two_values_missing_flag"] == 1]
```

```python
def single_fill_conditionally_with_geometric_average(t_0,
                                                     t_plus_1,
                                                     fund,
                                                     fund_returns,
                                                     index_returns,
                                                     fund_index_lookup):
    # if underlying is not traded, assume fund return is 0
    if index_returns.loc[t_0, fund_index_lookup[fund]] == 0:
        fund_returns.at[t_0, fund] = 0
    else:
        geometric_average = ((1 + fund_returns.loc[t_plus_1, fund]) ** (1/2)) - 1
        fund_returns.at[t_0, fund] = geometric_average
        fund_returns.at[t_plus_1, fund] = geometric_average

    return fund_returns


def double_fill_conditionally_with_geometric_average(t_0,
                                                     t_plus_1,
                                                     t_minus_1,
                                                     fund,
                                                     fund_returns,
                                                     index_returns,
                                                     fund_index_lookup):
    # if underlying is not traded, assume fund return is 0
    # stretch goal - TODO niche case handle where previous date
    # is traded but current date is not
    if index_returns.loc[t_0, fund_index_lookup[fund]] == 0:
        fund_returns.at[t_0, fund] = 0
        fund_returns.at[t_minus_1, fund] = 0
    else:
        geometric_average = ((1 + fund_returns.loc[t_plus_1, fund]) ** (1/3)) - 1
        fund_returns.at[t_minus_1, fund] = geometric_average
        fund_returns.at[t_0, fund] = geometric_average
        fund_returns.at[t_plus_1, fund] = geometric_average

    return fund_returns
```

```python
for date in dates_to_modify:
    for fund in fund_returns.columns:

        if fund_returns_copy.loc[date, "one_value_missing_flag"] == 1:
            fund_returns = single_fill_conditionally_with_geometric_average(
                t_0=date,
                t_plus_1=fund_returns_copy.loc[date, "t_plus_1"],
                fund=fund,
                fund_returns=fund_returns,
                index_returns=index_returns,
                fund_index_lookup=fund_index_lookup)

        elif fund_returns_copy.loc[date, "two_values_missing_flag"] == 1:
            fund_returns = double_fill_conditionally_with_geometric_average(
                t_0=date,
                t_plus_1=fund_returns_copy.loc[date, "t_plus_1"],
                t_minus_1=fund_returns_copy.loc[date, "t_minus_1"],
                fund=fund,
                fund_returns=fund_returns,
                index_returns=index_returns,
                fund_index_lookup=fund_index_lookup)
```

```python
# observing filled values - note that A I J K L are all canadian based, makes sense to be 0

fund_returns.loc[fund_returns_copy["two_values_missing_flag"] == 1]
```

```python
assert fund_returns.isna().any().sum() == 0
```

```python
# set the returns to our cleaned data

returns['Fund Total Returns (CAD)'] = fund_returns
```

```python
assert returns.isna().any().sum() == 0
```
