## Addressing holiday fund returns

It is likely that the reason for null values is due to the Canadian holiday schedule. A problem arises wherein if the underlying assets on which a fund is based are traded on the Canadian holiday (i.e. on foreign exchanges--consider US equities being traded on Canada day), then having a 0% return on the Canadian holiday will likely systemically over-represent tracking error for the fund.

We will accordingly use a two-pronged approach for filling missing values. We first do a check if the underlying was traded (non-zero return as a proxy). If not traded, we fill with 0. If traded, we will use a geometric average of the following day's trade and 0% return, such that the information of the fund movement is spread across the holiday and the following day.

Looking at the null holiday values, we see a situation where even though fund D, F, and fund J have values present, they are unlikely to be true values since they are either 0, very small values, or repeating very specific numbers. 

We will thus treat these funds as if they were null for these holiday values with respect to how we treat them from a null-filling standpoint.
